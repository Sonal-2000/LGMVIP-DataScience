```python
# import OpenCV library
from google.colab.patches import cv2_imshow
import cv2
```

```python
# Read the image
image = cv2.imread("rabit.jpg")
cv2_imshow(image)
cv2.waitKey(0)
```

```python
# Convert the original image to grayscale image
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_image)
cv2.waitKey(0)
```

```python
# Invert the grayscale image
inverted_image = 255 - gray_image
cv2_imshow(inverted_image)
cv2.waitKey()
```

```python
# blur the image by using the Gaussian Function in OpenCV 
blurred = cv2.GaussianBlur(inverted_image, (21, 21), 0)
cv2_imshow(blurred)
```

```python
# invert thr blurred image
inverted_blurred = 255 - blurred
pencil_sketch = cv2.divide(gray_image, inverted_blurred, scale=256.0)
cv2_imshow(pencil_sketch)
cv2.waitKey(0)
```

```python
# Show the original image and converted sketch image
cv2_imshow(image)
cv2_imshow(pencil_sketch)
cv2.waitKey(0)
```
